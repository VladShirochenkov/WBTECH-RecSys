## Проработка проблематики

### 1 Повышение уровня персонализации

Предсказание будущих взаимодействий пользователей с товарами позволяет создавать более точные и персонализированные рекомендации, что способствует улучшению пользовательского опыта. Персонализация помогает:

- Увеличить конверсию: Рекомендации, соответствующие интересам пользователя, с большей вероятностью приводят к покупке.
- Удержать пользователей: Персонализированный контент поддерживает интерес пользователей к платформе, снижая уровень оттока.
- Повысить лояльность: Пользователи ценят, когда сервис предлагает релевантные им товары, что способствует долговременным отношениям с брендом.

### 2 Оптимизация ассортимента и маркетинговых стратегий

Анализируя предсказанные интересы пользователей, бизнес может:

- Оптимизировать складские запасы: Понимание востребованности товаров помогает эффективно управлять запасами и снижать издержки.
- Сегментировать аудиторию: Разделение пользователей по предсказанным интересам позволяет проводить таргетированные маркетинговые кампании.
- Улучшить ценовую стратегию: Анализ предпочтений пользователей помогает в разработке динамического ценообразования.

### 3 Увеличение среднего чека и жизненной ценности клиента (CLV)

Более точные рекомендации могут привести к увеличению количества покупок на одного пользователя, что напрямую влияет на средний чек и общую жизненную ценность клиента для бизнеса.

### 4 Повышение эффективности рекламных кампаний

Используя предсказанные взаимодействия, бизнес может более эффективно распределять рекламные бюджеты, фокусируясь на наиболее перспективных товарах и пользователях.

## Формализация задачи

### Определение типа задачи

Данная задача относится к области рекомендательных систем и может быть формализована как задача предсказания следующего действия пользователя или предсказания следующего товара, с которым пользователь будет взаимодействовать. В зависимости от специфики задачи, это может быть:

- Задача классификации: Предсказать вероятность взаимодействия пользователя с каждым из возможных товаров.
- Задача ранжирования: Отсортировать товары по вероятности взаимодействия и рекомендовать самые релевантные.
- Задача последовательного прогнозирования: Учитывать последовательность взаимодействий пользователя для предсказания следующего шага.

### Входные данные

Исходя из предоставленных данных, входные данные это:

1. Контент карточек товара:
   - Текстовая информация: Название, описание, категории, характеристики.

2. Картинки:
   - Изображения товаров, которые могут быть использованы для извлечения визуальных признаков с помощью методов компьютерного зрения (например, сверточных нейронных сетей).

3. События по взаимодействиям пользователей с товарами (кликстрим):
   - Информация о действиях пользователей, а так же время действия.


### Выходные данные

Целью является предсказание следующих взаимодействий, поэтому выходными данными могут быть:

- Идентификаторы товаров: Список товаров, с которыми пользователь, вероятно, взаимодействует в ближайшем будущем.
- Вероятности взаимодействия: Скор для каждого товара, показывающий вероятность заинтересованности пользователя.
- Ранжированный список: Упорядоченные по релевантности рекомендации для пользователя.

### Подходы к решению

Для реализации задачи можно использовать различные подходы, включая, но не ограничиваясь:

1. Коллаборативная фильтрация:
   - User-based: Находит пользователей с похожими поведениями и рекомендует товары на основе их взаимодействий.
   - Item-based: Находит похожие товары на основе взаимодействий пользователей и рекомендует их.

2. Контентно-ориентированные методы:
   - Использует характеристики товаров (текст и изображения) для создания профилей и рекомендации схожих товаров.

3. Гибридные модели:
   - Комбинируют коллаборативную фильтрацию и контентные методы для улучшения качества рекомендаций.

4. Машинное обучение и нейронные сети:
   - Последовательные модели (RNN, LSTM, Transformer): Учитывают последовательность взаимодействий пользователя для предсказания следующего шага.
   - Сверточные нейронные сети (CNN): Для извлечения признаков из изображений товаров.
   - Мультимодальные модели: Объединяют текстовые и визуальные данные для более точного представления товаров.

### Формализация задачи

Целевая переменная (Target): Следующее взаимодействие пользователя с товаром.

Формально:

Для каждого пользователя \( u \) и временного шага \( t \):

- Входные данные:
  - История взаимодействий \( Hu = \{(t1, p1), (t2, p2), ..., (t{t'}, p{t'})\} \), где \( pi \) – товар.
  - Характеристики товаров \( P = \{p1, p2, ..., pn\} \) с их текстовыми и визуальными признаками.
  - Дополнительные данные о пользователе и контексте (если доступны).

- Выходные данные:
  - Предсказание следующего товара \( p{t'+1} \), с которым пользователь \( u \) взаимодействует.


### Метрики оценки модели

Для оценки качества модели рекомендуется использовать следующие метрики:

- Precision@K и Recall@K: Определяют долю релевантных товаров среди топ-K рекомендаций.
- MAP (Mean Average Precision): Среднее значение средней точности по всем пользователям.
- NDCG (Normalized Discounted Cumulative Gain): Учитывает порядок рекомендаций и релевантность.
- Hit Rate: Показывает, насколько часто релевантный товар присутствует в рекомендациях.
- AUC-ROC и AUC-PR: Для оценки вероятностных предсказаний.

### Этапы реализации

1. Сбор и подготовка данных:
   - Интеграция различных источников данных (текст, изображения, кликстрим).
   - Очистка и предобработка данных.

2. Извлечение признаков:
   - Текст: Векторизация с помощью TF-IDF, word embeddings или моделей трансформеров.
   - Изображения: Извлечение признаков с помощью CNN или использование предобученных моделей.
   - Взаимодействия: Создание последовательных данных, учитывающих временной порядок.

3. Выбор и обучение модели:
   - Выбор подходящей архитектуры.
   - Обучение модели на исторических данных с учетом временных зависимостей.

4. Оценка модели:
   - Разделение данных на тренировочные, валидационные и тестовые наборы.
   - Использование выбранных метрик для оценки качества предсказаний.

```python
def extract_charc(char_list):
    """
    Преобразует массив характеристик в словарь {charcName: charcValues}.
    """
    charc_dict = {}

    for char in char_list:
        name = char.get('charcName')
        values = char.get('charcValues')
        if name and values is not None:
            # Преобразуем numpy массив в список для удобства
            if isinstance(values, np.ndarray):
                values = values.tolist()
            charc_dict[name] = values
    return charc_dict
```

```python
# Примененим функцию к колонке characteristics
filtered_df_text['charc_dict'] = filtered_df_text['characteristics'].apply(extract_charc)

# Разворачиваем словари в отдельные столбцы
charc_df = filtered_df_text['charc_dict'].apply(pd.Series)
# Объединеним с исходным DataFrame
result_df = pd.concat([filtered_df_text.drop(['characteristics', 'charc_dict'], axis=1), charc_df], axis=1)
```

```python
def clean_colornames(color_list):
    """
    Функция для очистки списка цветов.
    Удаляет лишние кавычки, пробелы и приводит к нижнему регистру.
    Возвращает очищенный список или пустой список, если ввод некорректен.
    """
    if not isinstance(color_list, np.ndarray):
        return []

    cleaned = []
    for color in color_list:
        if isinstance(color, str):
            # Удаляем лишние кавычки и пробелы, приводим к нижнему регистру
            color_cleaned = color.strip().strip("'\"").lower()
            if color_cleaned:  # Добавляем только непустые строки
                cleaned.append(color_cleaned)
    return cleaned

# Применяем функцию очистки к столбцу colornames
result_df['cleaned_colornames'] = result_df['colornames'].apply(clean_colornames)
```

```python
def text_prep(text):
    # Переводим текст в нижний регистр

    # Убираем всё, что не русская буква, и заменяем на пробел
    text = re.sub('[^а-яё0-9]', ' ', text)

    # Заменяем множественные пробелы на один
    text = re.sub('\s+', ' ', text)

    # Тримминг
    text = text.strip()

    # Удаление стоп-слов
    tokens = text.split()
    tokens = [word for word in tokens if word not in stop_words]

    # Лемматизация
    lemmatized = [morph.parse(word)[0].normal_form for word in tokens]

    # Объединение обратно в строку
    cleaned_text = ' '.join(lemmatized)

    return cleaned_text
```

```python
import spacy
from tqdm import tqdm
import pandas as pd
import multiprocessing

# Загрузка модели SpaCy
nlp = spacy.load('ru_core_news_md', disable=['parser', 'ner', 'lemmatizer'])

# Функция для токенизации с использованием nlp.pipe
def tokenize_spacy(texts, nlp, batch_size=1000, n_process=4):
    tokens = []
    # Используем tqdm для отображения прогресса
    for doc in tqdm(nlp.pipe(texts, batch_size=batch_size, n_process=n_process, disable=["tagger"]), total=len(texts)):
        tokens.append([token.text for token in doc])
    return tokens

# Разделение данных на пустые и непустые тексты
empty_mask = result_df['cleaned_text'] == "<ПУСТО>"
non_empty_texts = result_df.loc[~empty_mask, 'cleaned_text'].tolist()

# Токенизация непустых текстов
tokenized_non_empty = tokenize_spacy(non_empty_texts, nlp, batch_size=1000, n_process=4)

# Инициализация столбца 'tokens' как пустого столбца с типом object
result_df['tokens'] = [[] for _ in range(len(result_df))]

# Присвоение токенов непустым текстам
result_df.loc[~empty_mask, 'tokens'] = pd.Series(tokenized_non_empty, index=result_df.loc[~empty_mask].index)
```

```python
tqdm.pandas()

# Функция для получения взвешенного эмбеддинга
def get_weighted_embedding(tokens, model, tfidf_weights):
    vector_size = model.vector_size
    emb = np.zeros(vector_size)
    weight_sum = 0.0
    for token in tokens:
        if token in model.wv and token in tfidf_weights:
            weight = tfidf_weights[token]
            emb += model.wv[token] * weight
            weight_sum += weight
    if weight_sum > 0:
        emb /= weight_sum
    return emb

# Применение функции к каждому документу
def compute_embedding(row, model, tfidf_df):
    tokens = row['tokens']
    tfidf_weights = tfidf_df.loc[row.name].to_dict()
    return get_weighted_embedding(tokens, model, tfidf_weights)

# Обработка применения функции с прогресс-баром
result_df['embedding'] = result_df.progress_apply(
    lambda row: compute_embedding(row, model, tfidf_df),
    axis=1
)
```

```python
def extract_materials(comp_list):
    """
    Извлекает уникальные названия материалов из списка строк,
    удаляя все символы кроме кириллицы и разбивая строки на отдельные элементы.

    Аргументы:
    comp_list (list): Список строк, содержащих материалы и их проценты.

    Возвращает:
    set: Множество уникальных названий материалов.
    """
    materials = set()
    # Регулярное выражение для поиска последовательностей кириллических символов
    cyrillic_pattern = re.compile(r'[А-Яа-яёЁ]+')

    for list_item in comp_list:
        # Если элемент списка пустой или не строка, пропускаем
        if not isinstance(list_item, str):
            continue
        # Находим все совпадения кириллических слов
        found_materials = cyrillic_pattern.findall(list_item)
        # Добавляем найденные материалы в множество (для уникальности)
        for material in found_materials:
            # Приводим к нижнему регистру и убираем лишние пробелы
            cleaned_material = material.strip().lower()
            if cleaned_material:
                materials.add(cleaned_material)
    return materials

# Применяем функцию к каждой строке столбца 'Состав' и создаём новый столбец 'Материалы'
result_df['Материалы'] = result_df['Состав'].apply(extract_materials)

# Инициализируем MultiLabelBinarizer
mlb = MultiLabelBinarizer()

binary_columns = pd.DataFrame(mlb.fit_transform(result_df['Материалы']),
                              columns=mlb.classes_,
                              index=result_df.index)
```

```python
def extract_OG_OT_OB(param_str):
    """
    Извлекает параметры ОГ, ОТ и ОБ из строки.

    Функция ищет три последовательных числовых значения, разделенных любыми нецифровыми символами.

    :param param_str: Строка, содержащая параметры в разных форматах.
    :return: Список из трёх целых чисел [ОГ, ОТ, ОБ].
    """
    if not isinstance(param_str, str):
        return [0, 0, 0]  # Возвращаем нули, если param_str не строка или None

    # Регулярное выражение для поиска трёх чисел, разделённых любыми нецифровыми символами
    pattern = re.compile(r'(\d+)\D+(\d+)\D+(\d+)')
    match = pattern.search(param_str)

    if match:
        og, ot, ob = match.groups()
        return [int(og), int(ot), int(ob)]
    else:
        return [0, 0, 0]  # Возвращаем нули, если паттерн не найден

# Применение функции к колонке 'param_str' и создание новых колонок
result_df[['Параметры ОГ', 'Параметры ОТ', 'Параметры ОБ']] = result_df['param_str'].apply(extract_OG_OT_OB).apply(pd.Series)
```

```python
IMAGE_SIZE = (256, 256)
BATCH_SIZE = 32 * strategy.num_replicas_in_sync
EMBEDDING_DIM = 256
NOISE_FACTOR = 0.1
EPOCHS = 50
```

```python
def load_and_preprocess_image(image_path, target_size=IMAGE_SIZE):
    """
    Загружает изображение, изменяет его размер и нормализует пиксели.
    """
    image = load_img(image_path, target_size=target_size)
    image = img_to_array(image)
    image = image.astype('float32') / 255.0
    return image

def add_noise(image, noise_factor=NOISE_FACTOR):
    """
    Добавляет гауссов шум к изображению и обрезает значения до [0,1].
    """
    noisy_image = image + noise_factor * np.random.normal(loc=0.0, scale=1.0, size=image.shape)
    noisy_image = np.clip(noisy_image, 0.0, 1.0)
    return noisy_image
```

```python
def preprocess_pair(image_path):
    """
    Создает пару (зашумленное изображение, чистое изображение) для автоэнкодера.
    """
    image = load_and_preprocess_image(image_path)
    noisy_image = add_noise(image)
    return noisy_image, image

def create_dataset(image_paths, batch_size=BATCH_SIZE, shuffle=True):
    """
    Создает tf.data.Dataset из списка путей к изображениям.
    """
    dataset = tf.data.Dataset.from_tensor_slices(image_paths)

    def _preprocess(path):
        noisy, clean = preprocess_pair(path.numpy().decode('utf-8'))
        return noisy, clean

    def tf_preprocess(path):
        noisy, clean = tf.py_function(_preprocess, [path], [tf.float32, tf.float32])
        noisy.set_shape([IMAGE_SIZE[0], IMAGE_SIZE[1], 3])
        clean.set_shape([IMAGE_SIZE[0], IMAGE_SIZE[1], 3])
        return noisy, clean

    if shuffle:
        dataset = dataset.shuffle(buffer_size=len(image_paths))

    dataset = dataset.map(tf_preprocess, num_parallel_calls=tf.data.AUTOTUNE)
    dataset = dataset.batch(batch_size)
    dataset = dataset.prefetch(tf.data.AUTOTUNE)

    return dataset

# Создание обучающего и валидационного датасетов
train_dataset = create_dataset(train_paths, batch_size=BATCH_SIZE, shuffle=True)
val_dataset = create_dataset(val_paths, batch_size=BATCH_SIZE, shuffle=False)
```

```python
def build_denoising_autoencoder(input_shape=(256, 256, 3), embedding_dim=256, noise_factor=0.1):
    """
    Строит денаузинговый автокодировщик с добавлением шума на входе.

    :param input_shape: Форма входных данных (по умолчанию 256x256x3).
    :param embedding_dim: Размерность эмбеддинга.
    :param noise_factor: Интенсивность шума.
    :return: Кортеж из моделей автокодировщика и энкодера.
    """
    # Энкодер
    encoder_input = layers.Input(shape=input_shape, name='encoder_input')

    # Добавление шума к входу

    noisy_input = layers.GaussianNoise(noise_factor, name='gaussian_noise')(encoder_input)

    x = layers.Conv2D(64, (3, 3), activation='relu', padding='same')(noisy_input)
    x = layers.MaxPooling2D((2, 2), padding='same')(x)  # 128x128

    x = layers.Conv2D(128, (3, 3), activation='relu', padding='same')(x)
    x = layers.MaxPooling2D((2, 2), padding='same')(x)  # 64x64
    x = layers.Conv2D(256, (3, 3), activation='relu', padding='same')(x)
    x = layers.MaxPooling2D((2, 2), padding='same')(x)  # 32x32
    x = layers.Conv2D(512, (3, 3), activation='relu', padding='same')(x)
    encoded = layers.MaxPooling2D((2, 2), padding='same', name='encoded')(x)  # 16x16

    # Сжимаем до эмбеддинга (в рамках EDA размер эмбеддинга будет 256)
    x = layers.Flatten()(encoded)
    embedding = layers.Dense(embedding_dim, activation='relu', name='embedding')(x)

    # Декодер
    x = layers.Dense(16 * 16 * 512, activation='relu')(embedding)
    x = layers.Reshape((16, 16, 512))(x)
    x = layers.Conv2DTranspose(256, (3, 3), strides=2, activation='relu', padding='same')(x)  # 32x32
    x = layers.Conv2DTranspose(128, (3, 3), strides=2, activation='relu', padding='same')(x)  # 64x64
    x = layers.Conv2DTranspose(64, (3, 3), strides=2, activation='relu', padding='same')(x)   # 128x128
    decoded = layers.Conv2DTranspose(3, (3, 3), strides=2, activation='sigmoid', padding='same', name='decoder_output')(x)  # 256x256

    # Модель автоэнкодера
    autoencoder = models.Model(inputs=encoder_input, outputs=decoded, name='denoising_autoencoder')

    # Модель энкодера для получения эмбеддингов
    encoder = models.Model(inputs=encoder_input, outputs=embedding, name='encoder')

    return autoencoder, encoder
```

```python
with strategy.scope():
    # Создание модели
    autoencoder, encoder = build_denoising_autoencoder(input_shape=(256, 256, 3), embedding_dim=EMBEDDING_DIM, noise_factor=NOISE_FACTOR)

    # Компиляция модели
    autoencoder.compile(optimizer=optimizers.Adam(), loss='mse', metrics=['mse'])
```

```python
from tensorflow.keras.callbacks import ReduceLROnPlateau

# Создание колбэка для уменьшения скорости обучения при отсутствии улучшений
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.8, patience=1, min_lr=1e-6, verbose=1)
# Определение путей для сохранения лучших весов
CHECKPOINT_PATH = 'best_autoencoder_weights.h5'
# Определение коллбеков
early_stop = callbacks.EarlyStopping(monitor='val_loss', patience=2, restore_best_weights=True, verbose=1)
model_checkpoint = callbacks.ModelCheckpoint(filepath=CHECKPOINT_PATH, monitor='val_loss', save_best_only=True, verbose=1)
```

```python
history = autoencoder.fit(
       train_dataset,
       epochs=EPOCHS,
       validation_data=val_dataset,
       callbacks=[early_stop, model_checkpoint, reduce_lr]
)
```

```python
AUTOTUNE = tf.data.AUTOTUNE
# Функция предобработки изображений
def preprocess_image(file_path):
    img = tf.io.read_file(file_path)
    img = tf.image.decode_jpeg(img, channels=3)
    img = tf.image.resize(img, IMAGE_SIZE)
    img = tf.cast(img, tf.float32) / 255.0  # Нормализация
    return img
# Создание полного набора данных для получения эмбеддингов
available_image_paths_tensor = tf.constant(available_image_paths)
full_dataset = tf.data.Dataset.from_tensor_slices(available_image_paths_tensor)
full_dataset = full_dataset.map(preprocess_image, num_parallel_calls=AUTOTUNE).batch(BATCH_SIZE).prefetch(AUTOTUNE)
```
